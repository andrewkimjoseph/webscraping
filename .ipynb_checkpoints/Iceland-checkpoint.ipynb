### 15. Project: Scrape Iceland tour data from Viator.com.

Number of Tours: 1139
List of URLs: https://www.viator.com/Iceland/d55-ttd

Tour Page Scrape:
- URL
- Name 
- The aggregate number of reviews
- Review rating
- Breadcrumb
- Price
- Overview
- What’s Included
- Departure & Return
- What to Expect
- Additional Info
- Cancellation Policy
- Traveler Photos
- Reviews

Tour Page Example: https://www.viator.com/tours/Reykjavik/Your-South-Coast-Full-Day-Tour-by-minibus/d905-36855P2

```python
import bs4
```

```python
import requests
```

```python
location_page = "https://www.viator.com/tours/Reykjavik/Blue-Lagoon/d905-105182P12"
```

```python
res = requests.get(location_page)
```

```python
soup = bs4.BeautifulSoup(res.text, 'lxml')
```

```python
print(soup.prettify())
```

```python
name = soup.find_all('h1')[0].text

name
```

```python
aggregate_reviews = soup.find("div", {"class": "ratingReviews__2CyF"}).text

aggregate_reviews
```

```python
rating = soup.find("span", {"class": "averageRatingValue__Q1ep"}).text

rating
```

```python
price = soup.find("span", {"class": "moneyView__2HPx"}).text

price
```

```python
overview_div = soup.find("div", {"class": "overviewWrapper__bMs4"})

overview = overview_div.get_text().replace('Overview', '')

overview
```

```python
whats_included = soup.find_all("div", {"class": "row__3-k3"})[6].text

whats_included
```

```python
departure_return = soup.find("div", {"class": "departureAndReturnContainer__3GNk"}).text

departure_return
```

```python
what_to_expect = soup.find("div", {"class": "introduction__2Nxq"})
```

```python
what_to_expect = soup.find("div", {"class": "productInfoCol__26F0"}).text
```

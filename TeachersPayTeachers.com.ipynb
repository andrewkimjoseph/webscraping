```python
from selenium import webdriver
import math
import itertools
import time
import csv
```

```python
def scrape():
    
    # CONFIGURING THE DRIVER
    path = "C:\webdrivers\chromedriver_win32\chromedriver.exe"
    agent = "user-agent=Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    headless = "--headless"
    the_options = webdriver.ChromeOptions()
    the_options.add_argument(path)
    the_options.add_argument(headless)
    
    # OPENING THE BROWSER
    driver = webdriver.Chrome(path, options=the_options)
    driver.maximize_window()
    
    # PREPPING THE OUTPUT FILE
    headers = ['Title', 'Price', 'Rating']
    table = [headers]
    
    current_page = 1
    while current_page < 42:
        
        url = f"https://www.teacherspayteachers.com/Browse/Page:{current_page}"
        driver.get(url)
        time.sleep(3)
        
        titles = driver.find_elements_by_xpath("//div[@class='SearchProductRowLayout']")
        prices = driver.find_elements_by_xpath("//div[@data-testid='productrow-price-original']")
        ratings = driver.find_elements_by_xpath("//div[@class='EvaluationStarRating__wrapper']/div")
        
        for (title, price, rating) in zip(titles, prices, ratings):
    
            title = title.find_element_by_xpath(".//h2").text
            price = price.text
            rating = "{:,}".format(math.trunc(float(rating.text.replace('k', '')) * 1000))
            record = [title, price, rating]
            table.append(record)
        
        current_page+=1
    
    return table
```

```python
materials = scrape()
```

```python
with open('teachers_csv.csv', mode = 'w', encoding = 'utf-8', newline = '') as teachers_csv:
    
    csv_writer = csv.writer(teachers_csv, delimiter = ',')

    csv_writer.writerows(materials)
```
